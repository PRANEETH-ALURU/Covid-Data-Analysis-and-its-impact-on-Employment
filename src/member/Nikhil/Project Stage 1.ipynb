```python
#copying the required features from the enrichment dataset to s1 variable
s1 = social_char_ER[["GEO_ID",
                     "DP02_0001E",
                     "DP02_0002E",
                     "DP02_0017E",
                     "DP02_0153E",
                     "DP02_0089E",
                     "DP02_0066E",
                     "DP02_0057E",
                     "DP02_0037E",
                     "DP02_0015E",
                     ]]


#Renaming the column id with name
s1 = s1.rename(columns={'DP02_0001E':'Total Households',
                        'DP02_0002E': 'Married-couple family', 
                        'DP02_0017E':'Average family size', 
                        'DP02_0153E':'Households with Internet subscription',
                        'DP02_0089E':'US Born Population', 
                        'DP02_0066E':'Population with Higher Education',
                        'DP02_0057E':'Number of Enrollments in High School',
                        'DP02_0037E':'Age of 15-50 Given Birth Last Year',
                        'DP02_0015E':'Households with Senior',
                        })
```

```python
#displaying the newly created sample datframe
s1.head()
```

### Splitting the GEO_ID into two in the social characteristics dataset 

```python
#spliting the GEO_ID column and considering the digits after 'US' in the GEO_ID 
s_temp = s1["GEO_ID"].str.split("US", n = 1, expand = True)
s1['countyFIPS']=s_temp[1]
```

```python
#displying top rows of s1
s1.head()
```

```python
#dropping the GEO_ID column in s1
s1_temp=s1.drop(labels='GEO_ID', axis=1)
s1_temp.head()
```

```python
#printing the type of the newly created coutyFIPS column
type(s1_temp.countyFIPS[1])
```

```python
#coverting the coutyFIPS datatype to int
#https://datatofish.com/string-to-integer-dataframe/
s1_temp['countyFIPS'] = pd.to_numeric(s1_temp['countyFIPS'])
type(s1_temp.countyFIPS[1])
```

```python
#copying the data from s1_temp to social_char
social_char_data=s1_temp.copy()
```

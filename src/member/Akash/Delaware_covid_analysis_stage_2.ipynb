# Analyzing the covid trends in Delaware (02/10/2022 - 02/17/2022)

```python
#importing the necessary libraries
import pandas as pd
import numpy as py
```

```python
# reading the Covid dataset
covid = pd.read_parquet("../../../data/covid_combined_dataset.parquet.gzip")
```

```python
covid
```

```python
Delaware = covid[covid['State'] == 'DE']
```

```python
Delaware
```

```python
#Taking the dates from 10th feb 2020
Status_DE = Delaware.loc[(Delaware["Date"] > "2022-02-09")]
```

```python
Status_DE.head()
```

```python
# Dropping 'countyFIPS','StateFIPS'
Status_DE = Status_DE.drop(['countyFIPS','StateFIPS'], axis=1)
```

```python
Status_DE
```

# Observation

 We are analyzing the Covid trends in Delaware form 10-feb-2022 to 17th feb 2022

## Cases

```python
# maximum number of Cases
Status_DE.Cases.max()
```

```python
# minimum number of Cases
Status_DE.Cases.min()
```

```python
# Average number of Cases
Status_DE.Cases.mean()
```

* As for the number of cases reported in Delware, the highest number of cases were reported on Nw castle county on 17th feb 2022 with '144131' cases

* Minimum were '48751' cases in Kent County on 10th feb 2022

* average reported cases between Feb 10th - Feb 17th 2022 were '84186'

## Deaths

```python
# maximum number of Deaths
Status_DE.Death.max()
```

```python
# minimum number of Deaths
Status_DE.Death.min()
```

```python
# Average number of Deaths
Status_DE.Death.mean()
```

* As for the number of deaths reported in Delware, the highest number of deaths were reported on Nw castle county on 17th feb 2022 with '1282' deaths

* Minimum were '576' deaths in Kent County on 10th feb 2022

* average reported deaths between Feb 10th - Feb 17th 2022 were '875'

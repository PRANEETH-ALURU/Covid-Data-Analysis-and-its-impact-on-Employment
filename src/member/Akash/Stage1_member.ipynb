### Enrichment Dataset (Reading and merging)

```python
# Importing the necessary libraries for this project
import pandas as pd
import numpy as np
```

```python
#reading the prsident candidate data
pres_county_cand = pd.read_csv("../../../data/member/akash/president_county_candidate.csv")

#reading the president county data
pres_county = pd.read_csv("../../../data/member/akash/president_county.csv")

#reading the president state data
pres_state = pd.read_csv("../../../data/member/akash/president_state.csv")
```

Renaming coulmn names for our easy understanding

```python
#renaming the total_votes to state_total_votes 
pres_state.rename(columns = {'total_votes' : 'state_total_votes'}, inplace = True)
pres_state.head()
```

```python
pres_county_cand.head()
```

```python
#renaming 'total votes' to 'cand_votes'
pres_county_cand.rename( columns = {'total_votes' : 'cand_votes'}, inplace = True)
pres_county_cand
```

```python
pres_county
```

```python
# merging the datasets 'pres_county' and 'pres_county_cand'
pres_merged = pd.merge(pres_county, pres_county_cand, on = ['state', 'county'])
```

```python
pres_merged
```

```python
# merging 'pres_merged' and 'pres_state'
pres_merged_final = pd.merge(pres_merged, pres_state, on = ['state'])
```

```python
pres_merged_final.head()
```

### Merging the enrichment dataset to Covid Dataset

##### Since the Covid dataset in Parquet format we need to install Pyarrow to access

```python
# reading the Covid dataset
covid = pd.read_parquet("../../../data/covid_combined_dataset.parquet.gzip")
```

```python
covid.head()
```

```python
#renaming 'County Name' to 'County_Name'
covid.rename ( columns = {'County Name' : 'County_Name'}, inplace = True)
```

```python
pres_merged_final.head()
```

```python
# renaming the coulmns in 'pres_merged_final' to match 'Covid dataset'
pres_merged_final.rename ( columns = {'county' : 'County_Name'} , inplace = True)
```

```python
pres_merged_final.drop(["state"], axis = 1, inplace = True)
```

```python
pres_merged_final.head()
```

```python
# we need to strip the data to remove the whitespace
covid['County_Name'] = covid['County_Name'].str.rstrip()
```

```python
covid[covid['County_Name'] == 'Kent County']
```

```python
covid.head()
```

Sampling the dataset as the sizes are larger

```python
# sampling the covid dataset
sample_covid = covid.sample(frac = 0.5, random_state = 1)
```

```python
sample_covid.head()
```

```python
# sampling the merged polls dataset
sample_pres_merged_final = pres_merged_final.sample(frac = 0.25, random_state = 1)
```

```python
sample_pres_merged_final.head()
```

```python
# merging the enrichment dataset to Covid dataset
merged_covid_polls = pd.merge (sample_pres_merged_final, sample_covid, on = ['County_Name'] )
```

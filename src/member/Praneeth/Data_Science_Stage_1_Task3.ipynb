### Plotting Trends of Covid Cases in the Last Week (North Carolina State)  {First Bullet}

```python
# Getting Covid Cases in NC for the last week of the data
super_covid1 = super_covid[super_covid['State']=='NC'] # considering only NC state data
super_covid1 = super_covid1.groupby(['Date'])['Cases'].sum() # adding all the cases by grouping with date
super_covid2 = super_covid1.reset_index().tail(7) # taking only last 7 days data
super_covid2
```

#### Displaying the plot

```python
# plotting the last week covid cases trend in NC
plt.figure(figsize=(11,7))
plt.plot(super_covid2['Date'],super_covid2['Cases'])
plt.title("LAST WEEK COVID CASES TREND IN NORTH CAROLINA")
plt.xlabel('Date')
plt.ylabel('Covid Cases')
```

**We can see that from 13th Feburary, 2022 the number of cases got soaring.**
